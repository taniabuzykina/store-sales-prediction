```python
data_train['Item_Fat_Content'].value_counts()
```

```python
#Replacing LF, low fat to Low Fat and reg to Regular
data_train['Item_Fat_Content'] = data_train['Item_Fat_Content'].replace({'LF': 'Low Fat', 'low fat': 'Low Fat', 'reg': 'Regular'})
data_test['Item_Fat_Content'] = data_test['Item_Fat_Content'].replace({'LF': 'Low Fat', 'low fat': 'Low Fat', 'reg': 'Regular'})

data_train['Item_Fat_Content'].value_counts()
```

```python
##HAndling categorical feature Item_Fat_Content
data_train['Item_Fat_Content']=data_train['Item_Fat_Content'].map({'Low Fat':1,'Regular':2})
data_test['Item_Fat_Content']=data_test['Item_Fat_Content'].map({'Low Fat':1,'Regular':2})
```

```python
#	Outlet_Location_Type
data_train['Outlet_Location_Type'].value_counts()
```

```python
##HAndling categorical feature Outlet_Location_Type
data_train['Outlet_Location_Type']=data_train['Outlet_Location_Type'].map({'Tier 1':1,'Tier 2':2,'Tier 3':3})
data_test['Outlet_Location_Type']=data_test['Outlet_Location_Type'].map({'Tier 1':1,'Tier 2':2,'Tier 3':3})
```

```python
#	Outlet_Type
data_train['Outlet_Type'].value_counts()
```

```python
##HAndling categorical feature Outlet_Location_Type
data_train['Outlet_Type']=data_train['Outlet_Type'].map({'Supermarket Type1':1,'Supermarket Type2':2,'Supermarket Type3':3,'Grocery Store':4})
data_test['Outlet_Type']=data_test['Outlet_Type'].map({'Supermarket Type1':1,'Supermarket Type2':2,'Supermarket Type3':3,'Grocery Store':4})
```

```python
data_train.head()
```

```python
data_train.info()
```

```python
data_train.describe()
```

```python
#Imputation for Item_weight with mean since no outlier
data_train['Item_Weight'].fillna(data_train['Item_Weight'].mean(),inplace=True)
data_test['Item_Weight'].fillna(data_test['Item_Weight'].mean(),inplace=True)
```

```python
data_train['Outlet_Size'].value_counts()
```

```python
#Imputation for Outlet_Size with mode as categorical variable
data_train.fillna(data_train['Outlet_Size'].mode()[0],inplace=True)
data_test.fillna(data_test['Outlet_Size'].mode()[0],inplace=True)
```

```python
#Handling categorical varialble Outlet_size
data_train['Outlet_Size']=data_train['Outlet_Size'].map({'Small':1,'Medium':2,'High':3})
data_test['Outlet_Size']=data_test['Outlet_Size'].map({'Small':1,'Medium':2,'High':3})
```

```python
data_train.isnull().sum()
```

```python
#Select required data fields
data_train.drop(['Item_Identifier','Outlet_Identifier'],axis=1,inplace=True)
data_test.drop(['Item_Identifier','Outlet_Identifier'],axis=1,inplace=True)
```
